```python
import requests
import pandas as pd

# Get company profiles
def get_company_info(symbol):
    token = <>
    url = f'https://financialmodelingprep.com/api/v3/profile/{symbol}?apikey={token}'
    response = requests.get(url).json()
    return pd.DataFrame(response)

# # Get balance shet
# def get_balance_sheet(symbol):
#     # quarterly available
#     token = <>
#     url = f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{symbol}?apikey={token}'
#     response = requests.get(url).json()
#     return pd.DataFrame(response)

# Get annual income statement
def get_income_statement(symbol, period='annual', limit=6):
    # period = annual or quarter
    # limit = number of period
    token = <>
    url = f'https://financialmodelingprep.com/api/v3/income-statement/{symbol}?period={period}&limit={limit}&apikey={token}'
    response = requests.get(url).json()

    return pd.DataFrame(response)[['date', 'symbol', 'revenue', 'grossProfit', 'ebitda']]

def get_ratio(symbol, period='annual', limit=6):
    # period = annual or quarter
    # limit = number of period
    token = <>
    url = f'https://financialmodelingprep.com/api/v3/ratios/{symbol}?period={period}&limit={limit}&apikey={token}'
    response = requests.get(url).json()
    return pd.DataFrame(response)[['symbol', 'date', 'returnOnCapitalEmployed']]

# Get enterprise value
def get_ev(symbol):
    # period = annual or quarter
    token = <>
    url = f'https://financialmodelingprep.com/api/v3/enterprise-values/AAPL?period=quarter&limit=6&&apikey={token}'
    response = requests.get(url).json()
    return pd.DataFrame(response)[['date', 'symbol', 'enterpriseValue']]
```

```python
company_info = pd.read_csv('../Company_Valuation/data/company_info.csv')
symbols = company_info['symbol']
symbols
```

```python
company_IS = []

for symbol in symbols:
    try:
        company_IS.append(get_income_statement(symbol))
    except:
        pass
    print(symbol)
```
